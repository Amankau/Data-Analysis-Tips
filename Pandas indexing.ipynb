## Conditions

```python
df.loc[(df.SepalLengthCm>6.5) & (df.Species == "Iris-virginica"),:]
```

```python
df.iloc[list((df.SepalLengthCm>6.5) & (df.Species == "Iris-virginica")),:] # using the list function to convert the output of multiple conditions into a boolean list
```

### loc and iloc are interchangeable when labels are 0-based integers
